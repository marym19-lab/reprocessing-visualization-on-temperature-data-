```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/temperatures.csv')
df
```

```python
df.isna().sum()
```

```python
df.ffill(inplace = True)
df.bfill( inplace=True)
```

```python
df['avg_temp_c'].fillna(df['avg_temp_c'].mean() , inplace=True)
```

```python
df.isna().sum()
```

```python
df.groupby('country')['avg_temp_c'].agg([max , min , sum])
```

```python
sns.boxplot(y='avg_temp_c' , x='country' , data=df , palette='crest')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.pointplot( x ='country',y='avg_temp_c'  , data=df , hue='avg_temp_c' , palette='crest')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.histplot(x='avg_temp_c' , bins=8 ,data=df )
```

```python
sns.barplot(x='country' , y='avg_temp_c' , data=df , palette='crest')
plt.xticks(rotation=90)
plt.show()
```
